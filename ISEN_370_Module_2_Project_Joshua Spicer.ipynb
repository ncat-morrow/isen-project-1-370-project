#Confidence Interval

Generate the 95% confidence interval for the mean estimated flow rate of Septemebr 2017.


````python
sep = df3["Est. Flow Rate[m^3/s]"].dropna()
sep = sep[sep >= 0].reset_index(drop=True)

octo = df4["Estimated Flow Rate (m3/s)"].dropna()
octo = octo[octo >= 0].reset_index(drop=True)

import scipy.stats as st
import numpy as np

mean_sep = np.mean(sep)
std_sep = np.std(sep, ddof=1)
n_sep = len(sep)

ci_sep = st.t.interval(0.95, df=n_sep-1, loc=mean_sep, scale=std_sep/np.sqrt(n_sep))
ci_sep
````

# Hypothesis testing
Suppose we are interested in testing if the mean estimated flow rate from September 2017 is statisticlly different from the mean estimated flow rate from October 2017.

***Question***<br>
How many samples are there in this experiment? What are they?<br>
***Answer***


````python
sep = df3["Est. Flow Rate[m^3/s]"].dropna()
sep = sep[sep >= 0].reset_index(drop=True)

octo = df4["Estimated Flow Rate (m3/s)"].dropna()
octo = octo[octo >= 0].reset_index(drop=True)
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(sep, octo, equal_var=False)
t_stat, p_value
````
