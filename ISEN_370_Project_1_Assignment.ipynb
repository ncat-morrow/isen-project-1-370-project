## Histogram Plotting
Plot a histogram for each of the three features of the dataset. Be sure to include proper titles and axis labels on each of your plots.

```python
import matplotlib.pyplot as plt
plt.hist(df["Heating Water Flow Rate"].dropna(), bins=30)
plt.title("Histogram: Heating Water Flow Rate")
plt.xlabel("Heating Water Flow Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df["Cooling Water Flow Rate"].dropna(), bins=30)
plt.title("Histogram: Cooling Water Flow Rate")
plt.xlabel("Cooling Water Flow Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df["TotalEnergy"].dropna(), bins=30)
plt.title("Histogram: TotalEnergy")
plt.xlabel("TotalEnergy")
plt.ylabel("Frequency")
plt.show()
```

## Scatter Plot

Create three scatter plots: the x axis for each plot is time and the y axis of the plot represents each of the three features.

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(df["Time"], df["Heating Water Flow Rate"], s=8)
plt.title("Time vs Heating Water Flow Rate")
plt.xlabel("Time"); plt.ylabel("Heating Water Flow Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(df["Time"], df["Cooling Water Flow Rate"], s=8)
plt.title("Time vs Cooling Water Flow Rate")
plt.xlabel("Time"); plt.ylabel("Cooling Water Flow Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(df["Time"], df["TotalEnergy"], s=8)
plt.title("Time vs TotalEnergy")
plt.xlabel("Time"); plt.ylabel("TotalEnergy")
plt.show()
```
