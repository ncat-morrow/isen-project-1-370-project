#Confidence Interval

Generate the 95% confidence interval for the mean estimated flow rate of Septemebr 2017.


````python
mean_sept = np.mean(df3['Est. Flow Rate[m^3/s]'])
````

````python
std_sept = np.std(df3['Est. Flow Rate[m^3/s]'], ddof=1)
````

````python
n_sept = len(df3['Est. Flow Rate[m^3/s]'])
````

````python
conf_int = stats.t.interval(0.95, n_sept-1, loc=mean_sept, scale=std_sept/np.sqrt(n_sept))
````

````python
conf_int
````

Interpret the confidence interval you generated.<br>
***Answer:***

The 95% confidence interval means we are 95% confident that the true average estimated flow rate in September 2017 lies between the lower and upper limits of the interval. If we repeated this study many times, about 95% of the intervals would contain the true mean.
